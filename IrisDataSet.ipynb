```python
import numpy as npy
import pandas as pd
import seaborn as sd
import matplotlib.pyplot as plt
```

```python
iris_set=sd.load_dataset('iris')
iris_set
```

```python
iris_set['species'].value_counts()
```

```python
iris_set['sepal_length'].value_counts()
```

```python
sd.scatterplot(iris_set['petal_width'],iris_set['sepal_length'],color='b',hue=iris_set['species'])
plt.show()
```

```python
sd.boxplot(iris_set['petal_width'],iris_set['sepal_length'])
plt.show()
```

```python
sd.violinplot(x=iris_set['sepal_length'],y=iris_set['species'])
plt.show()
```

```python
sd.jointplot(x=iris_set['sepal_length'],y=iris_set['sepal_width'],kind='hex')
plt.show()
```

```python
sd.jointplot(x=iris_set['sepal_length'],y=iris_set['sepal_width'],kind='kde')
plt.show()
```

```python
sd.jointplot(x=iris_set['sepal_length'],y=iris_set['sepal_width'],kind='reg')
plt.show()
```

```python
sd.jointplot(x=iris_set['sepal_length'],y=iris_set['sepal_width'],kind='resid')
plt.show()
```
